```python
import pandas as pd
```

```python
df = pd.read_csv('df.csv')
```

```python
df
```

```python
def dollar_K_remover(a):
  a = a.replace('$','')
  a = a.replace('K','')
  return(a)
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(dollar_K_remover)
```

```python
df = df[~df['Salary Estimate'].str.contains('Per Hour')]
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
def space_fix(a):
  a = a.replace('-',' - ')
  return(a)
```

```python
df['Salary Estimate'].apply(space_fix)
```

```python
df_new = df
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(space_fix)
```

```python
def splitter(a):
  a = a.split()
  return (a)
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(splitter)
```

```python
df['Salary Estimate'][0].split()
```

```python
df
```

```python
def min_finder(a):
  a = a[0]
  return(a)
```

```python
df['min_salary'] = df['Salary Estimate'].apply(min_finder)
```

```python
df.info()
```

```python
def max_finder(a):
  a = a[2]
  return(a)
```

```python
df['max_salary'] = df['Salary Estimate'].apply(max_finder)
```

```python
df.info()
```

```python
df['min_salary'] = df['min_salary'].astype('int')
```

```python
df['max_salary'] = df['max_salary'].astype('int')
```

```python
df[df['max_salary'] == '250(Employer']
```

```python
def employer_fix(a):
  a = a.replace('(Employer',' (Employer')
  a = a.split()[0]
  return(a)
```

```python
df['max_salary'] = df['max_salary'].apply(employer_fix)
```

```python
df.info()
```

```python
df['average_salary'] = (df['min_salary']+df['max_salary'])/2
```
