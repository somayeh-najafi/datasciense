# The usual pipeline for any images would be this:
## 1) Grey scale image
##2) Edge detection using A) Binary image through Tresholding B) Edge detection canny()
##3) Detecting Contours
##4) drawing contours

#First Method: Thresholding

```python


# Load the image

image = cv2.imread('vllCR.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply a threshold to get a binary image
_, thresh = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY_INV) #This simultaneousely inverts the colors

# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
image_contours = image.copy()
cv2.drawContours(image_contours, contours, -1, (0, 0, 0), 3)

# Display the original image and the image with contours
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Image with Contours")
plt.imshow(cv2.cvtColor(image_contours, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

# Print the number of contours found
print(f"Number of contours found: {len(contours)}")
```

# The second way using edge detection instead of threshold

```python

# Load the image
image = cv2.imread('vllCR.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply GaussianBlur to reduce noise and improve edge detection
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Apply Canny edge detector
edges = cv2.Canny(blurred, 30, 100)

# Find contours
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(len(contours))
# Draw contours on the original image
contour_image = image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

# Display the result
plt.imshow(contour_image[:,:,::-1])
cv2.waitKey(0)
cv2.destroyAllWindows()
```
