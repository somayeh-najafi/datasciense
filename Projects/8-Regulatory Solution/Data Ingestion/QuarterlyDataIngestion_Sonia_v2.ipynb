## Encoding categorical data to integer

```python
#new approach toward encoding

def encoding(item):
  if item == 'None' or item == 'Pass':
    return 1
  elif item == 'Minor' or item == 'Fail' or item == 'Within past year' or item == 'Flagged':
    return 2
  elif item == 'Major' or item == 'Within past 1-3 years' :
    return 3
  else:
    return 3
```

```python
for column in new_data.columns[1:]:
 new_data[column] = new_data[column].apply(encoding)
```

#Normalizing Data

```python
scaler = MinMaxScaler()
new_data_scaled=scaler.fit_transform(new_data)
cols=new_data.columns
new_data=pd.DataFrame(new_data_scaled,columns=cols)
```
