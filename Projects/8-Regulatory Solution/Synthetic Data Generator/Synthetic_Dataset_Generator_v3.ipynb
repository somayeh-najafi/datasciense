```python
import numpy as np
import pandas as pd
```

# Creating Synthetic Dataset

```python
num_samples=10
#np.random.seed(42)
Entity_id= np.random.randint(1,5001,num_samples)

data = {
    'entity': ["E"+str(num)for num in Entity_id],
    'annual_clients': np.random.randint(50,1000,num_samples),
    'infraction_type': np.random.choice(['None','Minor','Major'],num_samples),
    'infraction_timeline': np.random.choice(['None', 'Within past year', 'Within past 1-3 years'],num_samples),
    'public_complaints': np.random.choice(['None', 'Minor', 'Major'], num_samples),
    'sentiment_analysis': np.random.choice(['None', 'Flagged'], num_samples),
    'inspection_result': np.random.choice(['Pass', 'Fail', 'None'], num_samples)

}
```

```python
# Converting the dictionary to Pandas dataframe
df = pd.DataFrame(data)
```
