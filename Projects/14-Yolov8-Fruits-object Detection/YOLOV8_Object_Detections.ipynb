```python
from ultralytics import YOLO
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

```

```python
model_path = '/content/runs/detect/train2/weights/best.pt'  #if you want to use your trained model
image_path = '/content/apples-bananas.webp' #Test image
```

```python
img = cv2.imread(image_path)
```

```python
h,w,_ = img.shape
```

```python
model = YOLO(model_path) #Using our trained model
#model = YOLO("yolov8n.pt") #Using pre-trained model
```

#After prediction it saves the outpt in the runs/detect/predict folder

```python
#results = model(img)
results = model.predict(source=img,conf=0.25,save=True)
```

#Some extra steps

#Show output in Jupyter

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Assuming 'results' contains the prediction results
# and 'img' is the path to the original image

# Extract bounding boxes, class indices, and confidence scores
boxes = results[0].boxes.xyxy.cpu().numpy()  # Bounding box coordinates
class_indices = results[0].boxes.cls.cpu().numpy()  # Class indices
scores = results[0].boxes.conf.cpu().numpy()  # Confidence scores

# Mapping of class indices to class names
class_names = model.names

# Load the original image using OpenCV
#image = cv2.imread(img)

# Convert the image from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Set up a plot to display the image
fig, ax = plt.subplots(1, figsize=(12, 9))
ax.imshow(img)

# Iterate through detected boxes and draw them
for box, score, class_idx in zip(boxes, scores, class_indices):
    # Extract box coordinates
    xmin, ymin, xmax, ymax = box

    # Get the class name
    label = class_names[int(class_idx)]

    # Draw the rectangle around the object
    rect = patches.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin, linewidth=2,
                             edgecolor='red', facecolor='none')
    ax.add_patch(rect)

    # Display the label and confidence score
    plt.text(xmin, ymin, f'{label}: {score:.2f}', color='red', fontsize=12,
             bbox=dict(facecolor='white', alpha=0.5))

plt.axis('off')  # Turn off the axis
plt.show()  # Display the image with bounding boxes
```
